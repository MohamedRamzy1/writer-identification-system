```python
forms_list = list()
with open('data/ascii/forms.txt') as f:
    for i in range(16):
        next(f)
    for line in f:
        forms_list.append(line.split())
```

```python
print(f"Document ID : {forms_list[0][0]}")
print(f"Writer IF : {forms_list[0][1]}")
print(f"Number of Sentences : {forms_list[0][2]}")
if forms_list[0][3] == "prt":
    print("Some lines correctly segmented.")
else:
    print("All lines correctly segmented.")
print(f"{forms_list[0][5]} out of {forms_list[0][4]} lines are correctly segmented.")
print(f"{forms_list[0][7]} out of {forms_list[0][6]} words are correctly segmented.")
```

```python
writers_docs = dict()
for form in forms_list:
    if form[1] not in writers_docs.keys():
        writers_docs[form[1]] = [form[0]]
    else:
        writers_docs[form[1]].append(form[0])
```

```python
print(f"Number of writers : {len(writers_docs.keys())}")
```

```python
for writer in writers_docs.keys():
    print(f"Writer {writer} has {len(writers_docs[writer])} documents in the dataset.")
```

```python
writers_count = dict()
for writer in writers_docs.keys():
    if len(writers_docs[writer]) not in writers_count.keys():
        writers_count[len(writers_docs[writer])] = 1
    else:
        writers_count[len(writers_docs[writer])] += 1
```

```python
for count in writers_count.keys():
    print(f"Number of writers with {count} documents = {writers_count[count]}")
```
